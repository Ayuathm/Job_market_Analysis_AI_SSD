# 🧹 Job Post Text Cleaning Notebook

This notebook cleans extracted text from job descriptions for further analysis (e.g., skill extraction, title classification).

```python
# ✅ Step 1: Import libraries
import pandas as pd
import re

# 👇 Mount Google Drive if needed
# from google.colab import drive
# drive.mount('/content/drive')
```

```python
# ✅ Step 2: Load the extracted text CSV
file_path = '/content/extracted_job_texts(1).csv'  # 🔁 Replace if your file path is different
df = pd.read_csv(file_path)
df.dropna(subset=['text'], inplace=True)
df.head()
```

```python
# ✅ Step 3: Clean the job description text
def clean_text(text):
    text = re.sub(r'\n+', ' ', text)  # Remove newlines
    text = re.sub(r'\s+', ' ', text)  # Collapse multiple spaces
    text = re.sub(r'CamScanner', '', text, flags=re.IGNORECASE)  # Remove watermark
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)  # Remove non-ASCII characters
    return text.strip()

df['cleaned_text'] = df['text'].apply(clean_text)
df[['filename', 'cleaned_text']].head()
```

```python
# ✅ Step 4: Save the cleaned dataset
output_path = '/content/cleaned_job_texts.csv'
df.to_csv(output_path, index=False)
print(f"✅ Cleaned dataset saved to: {output_path}")
```

```python
# ✅ Step 2: Import libraries
import pandas as pd
import spacy
from keybert import KeyBERT
from tqdm import tqdm
tqdm.pandas()

nlp = spacy.load("en_core_web_lg")
kw_model = KeyBERT()
```

```python
# ✅ Step 3: Load cleaned job descriptions
file_path = '/content/cleaned_job_texts.csv'  # Replace if needed
df = pd.read_csv(file_path)
df.dropna(subset=['cleaned_text'], inplace=True)
df.head()
```

```python
# ✅ Step 4: Extract skills using KeyBERT
def extract_skills(text):
    keywords = kw_model.extract_keywords(text,
                                        keyphrase_ngram_range=(1, 3),
                                        stop_words='english',
                                        top_n=10)
    return [kw[0] for kw in keywords]

df['keybert_skills'] = df['cleaned_text'].progress_apply(extract_skills)
df[['filename', 'keybert_skills']].head()
```

```python
# ✅ Step 5: Optionally extract noun phrases using spaCy
def extract_noun_chunks(text):
    doc = nlp(text)
    return list(set([chunk.text.strip() for chunk in doc.noun_chunks if 3 < len(chunk.text) < 50]))

df['noun_chunks'] = df['cleaned_text'].progress_apply(extract_noun_chunks)
df[['filename', 'noun_chunks']].head()
```

```python
# ✅ Step 6: Save results to CSV
output_path = '/content/job_skills_extracted.csv'
df.to_csv(output_path, index=False)
print(f"✅ Skills saved to: {output_path}")
```

```python
# prompt: Add another column for job title and extract it from the text

# ✅ Step 7: Extract Job Title
# Simple approach: Look for common title patterns near the beginning of the text.
# This is a basic approach; more sophisticated title extraction might involve NLP models.
def extract_job_title(text):
    # Look for common patterns like 'Job Title:', 'Position:', etc., followed by the title
    # Or patterns at the beginning that look like titles
    patterns = [
        r"Job Title:\s*(.+)",
        r"Position:\s*(.+)",
        r"(.+?)(?:Job Description|Responsibilities|Requirements|Qualifications)", # Look for text before common sections
        r"^(.+?)(?:\n|\r|\.|$)" # Look for the first line
    ]
    for pattern in patterns:
        match = re.search(pattern, text, re.IGNORECASE)
        if match:
            # Take the first captured group and clean it
            title = match.group(1).strip()
            # Simple cleaning to remove trailing punctuation or keywords
            title = re.sub(r'[:,\.;\s]+$', '', title)
            # Limit the length to avoid capturing too much text
            if len(title) > 5 and len(title) < 100:
                return title
    return "Unknown Title" # Return a default if no pattern matches

df['job_title'] = df['cleaned_text'].apply(extract_job_title)
df[['filename', 'job_title', 'keybert_skills']].head()

# ✅ Step 8: Save results with job title to CSV
output_path_with_title = '/content/job_skills_and_titles_extracted.csv'
df.to_csv(output_path_with_title, index=False)
print(f"✅ Skills and titles saved to: {output_path_with_title}")
```
